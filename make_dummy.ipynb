```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker('ko_KR')
Faker.seed(42)

# 1. Customers 데이터 (1,000명)
customers = []
grades = ['01', '02', '03', '04'] # VIP, GOLD, SILVER, BRONZE 매핑용
for _ in range(1000):
    cust_id = f'C{str(_+1).zfill(5)}'
    customers.append({
        "cust_id": cust_id,
        "cust_nm": fake.name(),
        "age": random.randint(20, 70),
        "gender": random.choice(['남', '여']),
        "grade_cd": random.choice(grades),
        "join_date": fake.date_between(start_date='-5y', end_date='today')
    })
df_customers = pd.DataFrame(customers)

# 2. Products 정보 (Seeds 매핑 테스트용 고정 데이터)
prod_list = [
    ('0102001', '정기예금', 'SAVING'),
    ('0102002', '보통예금', 'SAVING'),
    ('0201001', '신용대출', 'LOAN'),
    ('0201002', '담보대출', 'LOAN')
]

# 3. Transactions 데이터 (10,000행)
transactions = []
for i in range(10000):
    p = random.choice(prod_list)
    transactions.append({
        "tx_id": i + 1,
        "cust_id": random.choice(df_customers['cust_id']),
        "prod_cd": p[0],
        "tx_type": random.choice(['IN', 'OUT']),
        "amt": random.randint(1000, 10000000),
        "tx_date": fake.date_time_between(start_date='-1y', end_date='now'),
        "curr_cd": 'KRW'
    })
df_transactions = pd.DataFrame(transactions)

# 4. Loans 데이터 (500행 - 일부 고객만 대출 보유)
loans = []
loan_targets = random.sample(list(df_customers['cust_id']), 500)
for i, c_id in enumerate(loan_targets):
    p = random.choice([p for p in prod_list if p[2] == 'LOAN'])
    loans.append({
        "loan_id": f'L{str(i+1).zfill(5)}',
        "cust_id": c_id,
        "loan_type": p[1],
        "loan_amt": random.randint(5000000, 100000000),
        "interest_rate": round(random.uniform(3.0, 8.5), 2),
        "start_date": fake.date_between(start_date='-3y', end_date='-1y'),
        "end_date": fake.date_between(start_date='today', end_date='+2y')
    })
df_loans = pd.DataFrame(loans)

# CSV 저장 (DBeaver에서 임포트용)
df_customers.to_csv('customers.csv', index=False)
df_transactions.to_csv('transactions.csv', index=False)
df_loans.to_csv('loans.csv', index=False)

print("CSV 파일 생성 완료!")
```
